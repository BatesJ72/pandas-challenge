```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# purchase_data.head()

# Change column header format
def clean_header(header_string):
    return header_string.lower().replace(" ", "_")


df = purchase_data.rename(columns=clean_header)

df.head()
```

## Player Count

* Display the total number of players


```python
# Assumption: Each SN represents a different player

player_count = len(df["sn"].unique())

# Create summary data frame
pd.DataFrame({"Total Players": [player_count]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of unique items
item_count = len(df["item_id"].unique())
# print(item_count)

# Avg purchase price
item_avg_price_v1 = df["price"].mean()
item_avg_price_v2 = "$" + str(round(item_avg_price_v1, 2))
# print(item_avg_price_v2)

# Total number of purchases
total_purchases = df["purchase_id"].max()
# print(total_purchases)

# Total Revenue
total_revenue = "$" + str(df["price"].sum())
# print(total_revenue)

# Create summary data frame
pd.DataFrame(
    {
        "Number of Unique Items": [item_count],
        "Average Purchase Price": [item_avg_price_v2],
        "Total Number of Purchases": [total_purchases],
        "Total Revenue": [total_revenue],
    }
)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Get distinct genders
genders = df["gender"].unique()
# print(genders)

# Total players
total_count = len(df["sn"].unique())
# print(total_count)

# Male players
male_count = len(df[df["gender"] == "Male"]["sn"].unique())
# print(male_count)

# Female players
female_count = len(df[df["gender"] == "Female"]["sn"].unique())
# print(female_count)

# Other players
other_count = len(df[df["gender"] == "Other / Non-Disclosed"]["sn"].unique())
# print(other_count)

# Data check
# total_count == female_count + male_count + other_count

# Summary dict
summary_dict = {
    "Total Count": [male_count, female_count, other_count],
    "Percentage of Players": [
        (str(round(male_count / total_count, 4) * 100) + "%"),
        (str(round(female_count / total_count, 4) * 100) + "%"),
        (str(round(other_count / total_count, 4) * 100) + "%"),
    ],
}
# print(summary_dict)

# Summary DataFrame
pd.DataFrame(summary_dict, index=["Male", "Female", "Other / Non-Disclosed"])
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase count by gender
purchase_count_1 = df.groupby(by=["gender"]).count()
purchase_count = purchase_count_1["purchase_id"]
# print(purchase_count)

# Avg purchase price by gender
purchase_price_avg_1 = df.groupby(by=["gender"]).mean()
purchase_price_avg = round(purchase_price_avg_1["price"], 2)
# print(purchase_price_avg)

# Avg purchase total by gender
purchase_price_sum_1 = df.groupby(by=["gender"]).sum()
purchase_price_sum = purchase_price_sum_1["price"]
# print(purchase_price_sum)

# Avg purchase total per person by gender
purchase_price_person_avg_1 = df.groupby(by=["sn", "gender"]).sum()
# print(purchase_price_person_avg_1)
purchase_price_person_avg_2 = purchase_price_person_avg_1.groupby(by=["gender"]).mean()
# print(purchase_price_person_avg_2)
purchase_price_person_avg = round(purchase_price_person_avg_2["price"], 2)
# print(purchase_price_person_avg)


# Summary data frame
pd.DataFrame(
    {
        "Purchase Count": purchase_count,
        "Average Purchase Price": purchase_price_avg,
        "Total Purchase Value": purchase_price_sum,
        "Average Total Purchase per Person": purchase_price_person_avg,
    }
)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create age goups and add to df
age_bins = pd.cut(
    df["age"],
    bins=[1, 9, 14, 19, 24, 29, 34, 39, 100],
    labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"],
)
df["age_group"] = age_bins
# df.head()

# Get total count by age group
total_count_1 = df[["sn", "age_group"]]
# print(total_count_1)
total_count = total_count_1.groupby(by=["age_group"]).nunique()
# print(total_count)

# Get percentages by age group
percentage_players = round((total_count / total_count.sum()) * 100, 2)
# print(percentage_players)


# Summary data frame
pd.DataFrame(
    {
        "Total Count": total_count["sn"],
        "Percentage of Players": percentage_players["sn"],
    }
)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase count by age
age_purchase_count_1 = df.groupby(by=["age_group"]).count()
age_purchase_count = age_purchase_count_1["purchase_id"]
# print(age_purchase_count)

# Avg purchase price by age
age_purchase_price_avg_1 = df.groupby(by=["age_group"]).mean()
age_purchase_price_avg = round(age_purchase_price_avg_1["price"], 2)
# print(age_purchase_price_avg)

# Avg purchase total by age
age_purchase_price_sum_1 = df.groupby(by=["age_group"]).sum()
age_purchase_price_sum = age_purchase_price_sum_1["price"]
# print(age_purchase_price_sum)

# Avg purchase total per person by age
age_purchase_price_person_avg_1 = df.groupby(by=["sn", "age_group"]).sum()
# print(age_purchase_price_person_avg_1)
age_purchase_price_person_avg_2 = age_purchase_price_person_avg_1.groupby(
    by=["age_group"]
).mean()
# print(age_purchase_price_person_avg_2)
age_purchase_price_person_avg = round(age_purchase_price_person_avg_2["price"], 2)
# print(age_purchase_price_person_avg)


# Summary data frame
pd.DataFrame(
    {
        "Purchase Count": age_purchase_count,
        "Average Purchase Price": age_purchase_price_avg,
        "Total Purchase Value": age_purchase_price_sum,
        "Average Total Purchase per Person": age_purchase_price_person_avg,
    }
)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Find the big spenders
total_spent = df.groupby(by=["sn"]).sum()
# print(total_spent)

# Data Check - how many unique people - does it match the number pulled in previous step?
# print(len(df["sn"].unique()))

# print(total_spent.sort_values("price", ascending=False))

# Top 5 biggest spenders
big_spender = total_spent.nlargest(5, "price")
# print(big_spender.rename(columns = {"price": "total_purchase_value"}))
big_spender.columns = ["purchase_id", "age", "item_id", "total_purchase_value"]
# print(big_spender)

# # Fix weird error
big1a = pd.merge(big_spender, total_spent[["price"]], how="inner", on="sn")
big1 = big1a[["price"]]
big1.columns = ["total_purchase_value"]
# print(big1)

# Count of Purchases
total_purchases = df.groupby(by=["sn"]).count()
# print(total_purchases)

big2 = pd.merge(
    big1[["total_purchase_value"]], total_purchases[["item_id"]], how="inner", on="sn",
)
big2.columns = ["total_purchase_value", "purchase_count"]
# print(big2)

# Avg Purchase Price
avg_purchase_price1 = df.groupby(by=["sn"]).mean()
avg_purchase_price = round(avg_purchase_price1, 2)
# print(avg_purchase_price)
big3 = pd.merge(big2, avg_purchase_price[["price"]], how="inner", on="sn")
big3.columns = ["total_purchase_value", "purchase_count", "average_purchase_price"]
# print(big3)

# Final dataframe
big_spenders_final = big3[
    ["purchase_count", "average_purchase_price", "total_purchase_value"]
].sort_values("total_purchase_value", ascending=False)
# print(big_spenders_final)

# # data check
# bigspender = df[df["sn"] == "Dyally87"]
# print(bigspender)


# Summary data frame
pd.DataFrame(big_spenders_final)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get top 5 items
top_5a = df.groupby(by=["item_id"]).count().sort_values("purchase_id", ascending=False)
# print(top_5a)
# top_5 = top_5a.loc[[1, 2, 3, 4, 5], ["purchase_id"]]
top_5 = top_5a.iloc[:5]
top_5.columns = [
    "purchase_count_top",
    "sn_top",
    "age_top",
    "gender_top",
    "item_name_top",
    "price_top",
    "age_group_top",
]
# print(top_5)

# Merge tables 1
df_nodupes = df[["item_name", "item_id", "price"]].drop_duplicates()
# print(df_nodupes)
top = pd.merge(
    top_5[["purchase_count_top"]],
    df_nodupes[["item_name", "item_id", "price"]],
    how="left",
    on="item_id",
)
top.columns = ["item_id_top", "purchase_count_top", "item_name_top", "price_top"]
# print(top)


# Total Purchase Value
purchase_value = top["purchase_count_top"] * top["price_top"]
top["purchase_value_top"] = purchase_value
print(top)

# d = {
#         "Item ID": top[item_id_top],
#         "Item Name": top[item_name_top],
#         "Purchase Count": top[purchase_count_top],
#         "Item Price": top[price_top],
#         "Total Purchase Value": top[purchase_value_top],
#     }


# Top Final


# Summary data frame
pd.DataFrame(top)
```
